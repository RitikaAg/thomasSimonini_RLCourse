```
import numpy as np,gym,random
```

```
env=gym.make("Taxi-v2")
env.render()
```

```
action_size=env.action_space.n
state_size=env.observation_space.n
qtable=np.zeros((state_size,action_size))
print(qtable)
```

```
t_episodes=50000
test_episodes=100
max_steps=99
alpha=.7
gamma=.618
ep=1
max_ep=1
d_rate=.01
min_ep=.01
```

```
for episode in range(t_episodes):
  state=env.reset()
  done=False
  step=0
  for step in range(max_steps):
    e_E_tradeoff=random.uniform(0,1)
    if(e_E_tradeoff>ep):
      action=np.argmax(qtable[state,:])
    else:
      action=env.action_space.sample()
    new_state,reward,done,info=env.step(action)
    qtable[state,action]=qtable[state,action]+alpha*(reward+gamma*np.max(qtable[new_state,:])-qtable[state,action])
    state=new_state
    if done==True:
      break
    episode+=1
    ep=min_ep+(max_ep-min_ep)*np.exp(-d_rate*ep)
    
    
```

```
env.reset()
rewards=[]
for episode in range(test_episodes):
  state=env.reset()
  done=False
  step=0
  t_rewards=0
  print("*********************")
  print("EPISODE",episode)
  for step in range(max_steps):
    env.render()
    action=np.argmax(qtable[state,:])
    new_state,reward,done,info=env.step(action)
    t_rewards+=reward
    if done:
      rewards.append(t_rewards)
      print("Score",t_rewards)
      break
    state=new_state
env.close()
print("ScoreOvertime"+str(sum(rewards)/test_episodes))
    
  
```
